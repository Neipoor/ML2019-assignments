# Review Report on "An innovative method in face recognition: critique of the paper - Eigenfaces vs. Fisherfaces: Recognition Using Class Specific Linear Projection"



## Introduction

The report is a critique of a research paper which introduces a new algorithm for face recognition. It summarises the content of the research, highlights the innovations and reviews the technical quality of the paper. Besides assessing the appropriateness of the algorithm in its proposed application, other possible applications are also listed. Following is a detailed discussion of potential further developments of the research. It ends with an appraisal of the presentation of the research paper.

## Content 

The research work presents Fisherface, a new algorithm for face recognition based on Fisher’s Linear Discriminant (FLD). It demonstrates the superiority of Fisherface in recognising faces under various illuminating conditions and expressions when compared with other known face recognition methods, namely correlation, Eigenface (with or without the first three principal components) and linear subspaces. It addresses the challenge frequently encountered in face recognition techniques - a change in illumination or expression can cause a significant variation to the image of an individual so that the recognition accuracy is impaired. Fisherface, as the authors propose, utilises a projection with directions nearly orthogonal to the within-class scatter and will maximise the ratio of between-class scatter to that of within-class scatter. This feature renders the algorithm insensitive to variations in lighting or facial expression and will produce well separated classes in a low-dimensional subspace, hence lowering the error rates in face recognition. 

To demonstrate the efficiency of Fisherface, the authors first compare the different methods in terms of their mathematical handling, with the advantages and limitations of each method discussed. A specific detailed comparison is given between Eigenfaces and Fisherface, which represent principal components analysis (PCA) and FLD respectively. FLD is shown to achieve greater between-class scatter, which simplifies classification.

Follows is a detailed presentation of the experiments, which are conducted under three different settings: variations in lighting directions; variations in lighting, expression, eye wear; presence or absence of glasses. Fisherface is demonstrated to outperform other methods under all conditions.

## Innovation

As mentioned above, the main contribution of this paper is a new algorithm for face recognition. It is innovative in the following ways.

The most appealing innovation of the method is its insensitivity to large variations in lighting and facial expressions. As of the time of writing, there was still a lack of reliable face recognition techniques under extreme lighting conditions. Fisherface discounts the areas of face which tend to create self-shadowing or large variations between different images of the same individual, while retaining the effects of the areas more significant for recognition. This is achieved by projecting the images to a significantly lower dimensional feature space and choosing projection directions that are nearly orthogonal to the within-class scatter. Different from principal components analysis (PCA) which maximise the total scatter across all classes, Fisherface maximises the ratio of between-class scatter to that of within-class scatter. Not only does it negate the impact of lighting in this way, Fisherface can also be used to pick up other given traits on the face, while ignoring undesired variations. To put it another way, rather than utilising evenly every pixel of an image (as applied in some other methods), Fisherface places more weight on the parts of the face which are less affected by lighting and expressions. This feature enables identifying key facial characteristics even under great variations in lighting and facial expressions, which was achieved less satisfactorily using other methods mentioned. 

The research is also innovative in the fundamentals that the new algorithm is based upon.  it utilises the observation that for a Lambertian surface without shadowing, the images of a particular face lie in a 3D linear space of the high-dimensional image space. This observation enables the Linear Discriminant Analysis (LDA) to be applied for dimensionality reduction, after which the linear separability can still be preserved, and better between-class discrimination achieved. Neither correlation nor Eigenface utilises this observation. Second, Fisherface uses class specific linear methods for dimensionality reduction and simple classifiers in the reduced feature space. This achieved better recognition rates. 

## Technical quality

The research is of satisfactory technical quality. There are a few areas where a high level of research standards is demonstrated. 

Firstly, the results of the research will be easy to replicate as the authors gave a detailed presentation of the database used, the processes where the training and testing images were generated and the complete mathematical method of handling of the data. 

Secondly, the database chosen also suits the purpose of the research. The training/testing images in the first experiment are from Harvard Robotics Laboratory. These images were produced under strictly controlled lighting conditions, with the light source directions varying from 0 to 75 degrees. The advantage of using this database is the ease to categorise the images according to the deviation extent of the light source, which is almost impossible to achieve using random images from other sources. It also ensures to include images with a wide range of lighting directions. 

Thirdly, while comparing the performance of different methods on a given set of training/testing images, the authors made relevant adjustments to the images according to the method used to ensure the best performance of each method. This has ensured an impartial comparison between Fisherface and other methods. 

Nevertheless, the study is not without limitations. A very apparent drawback of the research is its sample size. In the first experiment, the training and testing samples consist of 330 images of only 5 people. As mentioned in (Martínez & Kak 2001), the size of the samples is likely to affect the performance of the analysis method of interest. Furthermore, with a group of 5 randomly chosen people, it is reasonable to assume there is a low possibility that any of them bear significant resemblance, which would presumably make the discrimination process easier. On the contrary, with a much larger sample size and a subsequently higher chance of two individuals resembling each other, it is intuitive to speculate there might be a greater possibility of mistakenning an individual with his/her look alike. This limitation is so obvious that the authors have themselves recognised it and precluded the possibility to extrapolate the results of the research to much larger databases.

Another limitation of the research is the insufficiency of the testing data in the first experiment (variation in lighting). In other words, the testing groups the authors have chosen to have represented only a part, but not a full spectrum of all possible lighting conditions. In the extrapolating experiment, images in subset 1 were used as the training set while testing samples were chosen from subsets 1 to 3. The proposed Fisherface demonstrated excellent performance, with an error rate of no more than 5%. However, given that images in these subsets have a light source which deviates within 45 degrees from the camera axis, the correspondent portraits are virtually free of any significant self-shadowing. This inevitably makes readers wonder if the high performance of Fisherface can be replicated under more extreme lighting conditions. Such query would have been easy to be addressed by including testing samples from subsets 4 and 5, which represent much more variable lighting directions. To have included samples with extreme lighting conditions might have made the results less ideal but would have improved the overall quality, as the authors did claim their method to be ‘insensitive to large variation in lighting direction’. It is worth mentioning that the authors do point out these methods may fail under extreme lighting conditions. Still, it would have been prudent to demonstrate this claim with real data from the experiment. 

## Application and X-factor

The proposed algorithm, Fisherface, is appropriate and indeed quite specific for facial recognition. It is based on the observation that certain parts of the face are less liable to the effect of various lighting conditions or expressions. To discount these parts is a clever thought, which reflect that fact that the same person can appear dramatically different under different lighting conditions.  

As Fisherface is based on Fisher’s Linear Discrimination, which is a ‘classical’ pattern recognition technique (Duda & Hart 1973), it can be presumed that the same principles can be applied to a number of various recognition tasks, such as generic object recognition and gesture discrimination.  It can also be incorporated into industrial robotics for executing certain tasks.  

More specifically, as the authors also pointed out in 3.3 Glasses Recognition, the Fisherface method can be used to categorize images according to other traits of images (including portraits). It was already demonstrated in the paper that the Fisherface can discriminate photos of people with or without glasses. Thus, it is reasonable to assume that the Fisherface method can also pick up other characteristics, such as the colour of hair and eyes, the shape of the nose and mouth, etc. Such abilities may be used in genetic studies or census of population. 

The research has displayed impressive results, but the results it presents has yielded a few questions that are worth exploring and can be served as interests of future research. The first question is regarding lighting and pose. As mentioned above, the testing samples in this research excluded the ones with extreme lighting conditions. Further complementary experiments can be done to investigate the performance of Fisherface in all possible illuminating situations. Similarly, pose variations should also be considered in the evaluation of a face recognition algorithm. Fisherface could be combined with other methods to handle pose variations of an individual, such as a template-based system (Beymer 1994) or by analysing the direction of gaze (Gee & Cipolla 1994). Afterall, it is impossible to guarantee all photos in real life have an almost vertical lighting and the person facing the camera.

The results of the interpolation experiment were also worth exploring further. In this experiment, samples from subsets 1 and 5 were used for training. Subsets 2-4 were used as the testing group. Interestingly, all the methods, except Fisherface, showed the worst performance (the highest error rates) in subset 3. On the contrary, the error rates were much lower in subset 4, the photos from which have greater self-shadowing which covers more area of the face. This is contradictory to the intuition that more self-shadowing would result in worse performance of these algorithms. One of the possible explanations is that the accuracy of face recognition is affected by the resemblance between training samples and testing samples. In other words, it is easier to recognise faces with similar illuminating conditions as those in the training samples. The samples from subset 4, while having greater self-shadowing, do bear more resemblance to the lighting conditions of the testing samples from subset 5 when compared to subset 3. To test this hypothesis, samples of a certain lighting direction can be used for training, while testing samples can be grouped according to the resemblance of their lighting conditions to the testing samples. The experiment can be repeated with different lighting conditions in the training group. 

If the above can be confirmed, then it brings forward another question: should we include as many as possible photos of varying illuminating conditions in the training set? Would this make the projection more accurate? More research can be done to investigate this possibility. These questions listed above will be interesting to be discussed in class. The most intriguing part of the work is that it demonstrates how different methods and information can affect the accuracy of machine learning, and the truth that meaningful information is far more significant than the mere amount of information.
